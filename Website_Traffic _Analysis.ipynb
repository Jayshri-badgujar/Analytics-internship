```python
import pandas as pd
import matplotlib.pyplot as plt

# Load data
df = pd.read_csv("traffic.csv")

# Clean column names
df.columns = df.columns.str.strip().str.lower().str.replace(" ", "_")

# Convert date
df['date'] = pd.to_datetime(df['date'], errors='coerce')

#  KPIs
total_events = len(df)
unique_artists = df['artist'].nunique()
unique_tracks = df['track'].nunique()
unique_albums = df['album'].nunique()
unique_cities = df['city'].nunique()

print("Total Events:", total_events)
print("Unique Artists:", unique_artists)
print("Unique Tracks:", unique_tracks)
print("Unique Albums:", unique_albums)
print("Unique Cities:", unique_cities)

# TOP TRACKS
top_tracks = df['track'].value_counts().head(10)
top_tracks.plot(kind='bar', title='Top Tracks')
plt.xticks(rotation=45)
plt.show()

# TOP ARTISTS 
top_artists = df['artist'].value_counts().head(10)
top_artists.plot(kind='bar', title='Top Artists')
plt.xticks(rotation=45)
plt.show()

# TOP ALBUMS 
top_albums = df['album'].value_counts().head(10)
top_albums.plot(kind='bar', title='Top Albums')
plt.xticks(rotation=45)
plt.show()

# EVENTS OVER TIME
events_over_time = df.groupby(df['date'].dt.date).size()
events_over_time.plot(title='Events Over Time')
plt.xticks(rotation=45)
plt.show()

# TOP CITIES
top_cities = df['city'].value_counts().head(10)
top_cities.plot(kind='bar', title='Top Cities')
plt.xticks(rotation=45)
plt.show()

print("✅ Analysis Complete!")
```
